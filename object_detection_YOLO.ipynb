```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the original and grayscale images
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.show()
```

```python
new_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
"""
1. Splitting images channels
"""
r,g,b = cv2.split(new_image) # we don't have any more channels
print('r', r.shape)
print('g', g.shape)
print('b', b.shape)
plt.subplot(1, 3, 1)
plt.imshow(g)
plt.subplot(1, 3, 2)
plt.imshow(g)
plt.subplot(1, 3, 3)
plt.imshow(b)
plt.show()
new_image = cv2.merge((r,g,b))
print(new_image.shape)

```

```python
""" 
2. Resizing of Image
"""
s = 10 
w = int(new_image.shape[1]*s/100)
h = int(new_image.shape[0]*s/100)
dim = (w,h)
re_size = cv2.resize(new_image, dim, interpolation = cv2.INTER_AREA)
print(re_size.shape)
plt.imshow(re_size)
```

```python
"""
3. Rotate Operation 
"""
(h, w) = new_image.shape[0:2] #taking hieght and width

c = (w/2, h/2) # finding the center

angle = 90

m = cv2.getRotationMatrix2D(c, angle, 1.0)
rotate_90 = cv2.warpAffine(new_image, m, (h,w))
plt.imshow(rotate_90)
```
